```python
category = os.listdir('data')[:3]
splits = os.listdir('data')[3:]
```

```python
for cat in category:
    data = glob(f'data/{cat}/*.jpg')
    random.shuffle(data)

    # train, valid, test 6:2:2
    for i in range(len(data)):
        if i < len(data)*0.6:
            shutil.move(data[i], f'data/train/{cat}')
        elif i < len(data)*0.8:
            shutil.move(data[i], f'data/valid/{cat}')
        else:
            shutil.move(data[i], f'data/test/{cat}')
```

```python
for cat in category:
    os.rmdir(f'data/{cat}')
```

```python
train_datagen = ImageDataGenerator(rescale=1./255,
                                   rotation_range=20,
                                   width_shift_range=0.05,
                                   height_shift_range=0.05,
                                   brightness_range=[0.2, 1.0])
```

```python
train_generator = train_datagen.flow_from_directory('data/train',
                                                    target_size=(250, 250),
                                                    batch_size=10,
                                                    class_mode='categorical')
```

```python
valid_datagen = ImageDataGenerator(rescale=1./255)
valid_generator = valid_datagen.flow_from_directory('data/valid',
                                                    target_size=(250, 250),
                                                    batch_size=10,
                                                    class_mode='categorical')
```

```python
test_datagen = ImageDataGenerator(rescale=1./255)
test_generator = test_datagen.flow_from_directory('data/test',
                                                    target_size=(250, 250),
                                                    batch_size=10,
                                                    class_mode='categorical')
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dropout, Flatten, Dense, Activation, BatchNormalization
from keras.callbacks import EarlyStopping, ReduceLROnPlateau
```

```python
model = Sequential()
model.add(Conv2D(16, (3, 3), input_shape=(250, 250, 3), activation="relu"))
model.add(MaxPooling2D(pool_size = (2, 2)))
model.add(Conv2D(32, (3, 3), activation="relu"))
model.add(MaxPooling2D(pool_size = (2, 2)))
model.add(Conv2D(64, (3, 3), activation="relu"))
model.add(MaxPooling2D(pool_size = (2, 2)))
model.add(Flatten())
model.add(Dropout(0.2))
model.add(Dense(128, activation="relu"))
model.add(Dropout(0.2))
model.add(Dense(3, activation="softmax"))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

model.summary()
```

```python
earlystop = EarlyStopping(patience=10) # 10번이상 모델이 개선되지 않으면 멈춤
```

```python
learning_rate_reduction = ReduceLROnPlateau(monitor='val_accuracy', 
                                            patience=2, 
                                            verbose=1, 
                                            factor=0.5, 
                                            min_lr=0.00001)  # learning_rate의 하한값
```

```python
callbacks = [earlystop, learning_rate_reduction]
```

```python
epochs = 100
batch_size = 10
```

```python
history = model.fit_generator(
    train_generator,
    epochs=epochs,
    validation_data=valid_generator,
    validation_steps=160//10,
    steps_per_epoch=480//10,
    callbacks=callbacks
)
```

```python
model.save("./cnn_model_current1.h5")
```
